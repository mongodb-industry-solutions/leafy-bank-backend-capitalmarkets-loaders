```python
# Import libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import warnings
warnings.filterwarnings("ignore")
```

```python
import datetime as dt

start_date = dt.datetime.strptime("2025-01-10", "%Y-%m-%d").date()
end_date = dt.datetime.strptime("2025-02-10", "%Y-%m-%d").date()
```

```python
# Retrieve the S&P 500 data for the past month with daily intervals
ticker = '^GSPC'  # S&P 500 ticker symbol
stock_data = yf.Ticker(ticker)
sp500 = stock_data.history(interval='1d', start=start_date, end=end_date)
```

```python
# Drop the 'Dividends' and 'Stock Splits' columns
sp500 = sp500.drop(columns=['Dividends', 'Stock Splits'])
```

```python
sp500.to_csv('sp500_data.csv')
```

```python
sp500.sort_values(by='Date', ascending=True, inplace=True)
sp500.tail(30)
```

```python
# Calculate the daily percentage change
sp500['Pct_Change'] = sp500['Close'].pct_change() * 100

# Filter for days with a drop of 5% or more
significant_drops = sp500[sp500['Pct_Change'] <= -1]

# Display the dates and percentage changes of significant drops
print(significant_drops[['Close', 'Pct_Change']])
```
